```python
import pandas as pd
import altair as alt
alt.data_transformers.enable("vegafusion")

customers_df = pd.read_csv(
    "/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/coding-problems/noahs-customers.csv")
#print(customers_df)

order_items_df = pd.read_csv(
    "/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/coding-problems/noahs-orders_items.csv")
#print(order_items_df)

orders_df = pd.read_csv(
    "/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/coding-problems/noahs-orders.csv")
#print(orders_df)

products_df = pd.read_csv(
    "/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/coding-problems/noahs-products.csv")
#print(products_df)
```

## We need to find a person who was:
- Born in either 1939, 1951, 1963, 1973, 1987, 1999
- Born between june 21 - july 22
- bought a bag from noahs

```python
rabbit_years = [1939, 1951, 1963, 1975, 1987, 1999]

# Cancer = June 21 to July 22
is_cancer = (
    ((customers_df['birthdate'].dt.month == 6) & (customers_df['birthdate'].dt.day >= 21)) |
    ((customers_df['birthdate'].dt.month == 7)
     & (customers_df['birthdate'].dt.day <= 22))
)

# Rabbit year
is_rabbit = customers_df['birthdate'].dt.year.isin(rabbit_years)

# Bought something from Noah's = appears in orders table
has_bought_from_noahs = customers_df['customerid'].isin(
    orders_df['customerid'])

candidates = customers_df[is_rabbit & is_cancer & has_bought_from_noahs].copy()

candidates[['customerid', 'name', 'birthdate',
            'phone', 'address', 'citystatezip']]

candidate_orders = (
    candidates
    .merge(orders_df, on='customerid')
    .merge(order_items_df, on='orderid')
    .merge(products_df, on='sku')
)

rug_cleaner_bought = (candidate_orders['desc'].str.contains('rug cleaner', case=False)).any()

candidate_bought_rug = candidate_orders[rug_cleaner_bought].copy()
```
