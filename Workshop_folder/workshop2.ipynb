```python
# Converting Catergories to Category Id
import pandas as pd
from redshift_connector import connect


def load_events(filename: str) -> list:
    return pd.read_csv("/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/Workshop_folder/events/new_events_1.csv", sep="\t", header=None)


def add_column_names(df: pd.Dataframe) -> pd.Dataframes:
    df.columns = ["id", "eventname", "dateid", "eventid",
                  "categoryid", "duration", "eventstart"]
    df["eventstart"] = pd.to_datetime(df['eventstart'])
    return df


def get_db_connection():
    return connect(host="c23-workgroup.129033205317.eu-west-2.redshift-serverless.amazonaws.com",
                   user="coach_chris",
                   password="Scoach_71",
                   database="dw_air_travel",
                   port=5439)


def get_categories(connection) -> list[dict]:
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM ticket.category;")
    data = cursor.fetchall()
    cursor.close()

    mapping = {}

    for line in data:
        print(line)
        mapping[line[1].strip()] = line[0]

    return mapping


def map_to_name(id, mapping):
    if id.isnumeric():
        return id

    return mapping[id]


def convert_category_name_to_id(df: pd.DataFrame, mapping: dict) -> pd.DataFrame:

    df["categoryid"] = df["categoryid"].apply(
        lambda id: map_to_name(id, mapping))

    return df


def upload_data(df: pd.DataFrame, connection) -> None:
    cursor = connection.cursor()

    insert_query = """
        INSERT INTO ticket.event (
            eventid, dateid, eventname, eventstart, cancelled, 
            categoryid
        )
        VALUES (%s, %s, %s, %s, %s, %s)
    """

    for _, row in df.iterrows():

        if pd.isna(row["eventstart"]):
            print(f"Skipping row with invalid eventstart: {row.to_dict()}")
            continue

        cursor.execute(insert_query, (
            int(row["eventid"]),
            int(row["dateid"]),
            row["eventname"],
            row["eventstart"],
            False,
            int(row["categoryid"])
        ))

    connection.commit()
    cursor.close()



if __name__ == "__main__":
    df = load_events("new_events_1.csv")
    df = add_column_names(df)
    conn = get_db_connection()
    mapping = get_categories(conn)
    df = convert_category_name_to_id(df, mapping)
    print(df)
    upload_data(df, conn)
    print("Upload complete!")
    conn.close()
```
