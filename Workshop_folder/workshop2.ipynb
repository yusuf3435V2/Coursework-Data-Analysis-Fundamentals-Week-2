```python
import pandas as pd 
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
sharks = pd.read_csv(
    "/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/cleaner_sharks.csv")
print(sharks)
```

```python
alt.Chart(sharks).mark_bar().encode( 
    x=alt.X("type", title="Type of incident", sort="-y"), 
    y=alt.Y("count()", title="Count"),
    color=alt.Color('type', legend=None)
    ).properties(
        title="Distribution of Incidents by Type",
        width=600,
        height=400)
```

```python
alt.Chart(sharks).mark_bar().encode(
    x=alt.X("type", title="Type of incident", sort="-y"),
    y=alt.Y("count()", title="Count"),
    color=alt.Color('fatal')
).properties(
    title="Fatality by reason",
    width=600,
    height=400)

```

```python
top_countries = sharks["country"].value_counts().head(10).index

filtered_sharks = sharks[sharks['country'].isin(top_countries)]

alt.Chart(filtered_sharks).mark_bar().encode(
    x = alt.X("country", sort="-y"),
    y = alt.Y("count()"),
    color="sex"
).properties(width=500, height=400)
```

```python
alt.Chart(sharks).mark_bar().encode(
    x=alt.X("type", title= "Type of attack", sort="-y"),
    y=alt.Y("count()", title= "number of incidents"),
    color=alt.Color("fatal", title="Was Fatal?"),
    tooltip=[

        alt.Tooltip("type", title='Type of Attack'),
        alt.Tooltip("fatal", title='Was Fatal?'), 
        
    ]

).properties( 
    width=800, height=400, title="Shark Incidents Over Time by Type"
)

```

```python
library = pd.read_csv("/Users/yusufahmed/Documents/PreCourseLearning/week8_data_analysis_2/Workshop/library_users_clean.csv")
library.info()

```

```python
alt.Chart(library).mark_bar().encode(
    x=alt.X("Patron Type Definition", title="Type of Patron", sort="-y"),
    y=alt.Y("count()", title="number of patrons"),
    tooltip=[

        alt.Tooltip("Patron Type Code", title='Type of patron'),
        alt.Tooltip("count()", title='count'),

    ]
).properties(
    width=800, height=400, title="Patron type and number of patrons"
)
```

```python
alt.Chart(library).mark_point().encode(
    x=alt.X("Total Checkouts", title="Total Checkouts", sort="-y"),
    y=alt.Y("Total Renewals", title="Total Renewals"),
    tooltip=[

        alt.Tooltip("Total Checkouts", title='Type of patron'),
        alt.Tooltip("Total Renewals", title='count'),

    ]
).properties(
    width=800, height=400, title="Patron type and number of patrons"
)
```
