```python
# You can also chain functions together
shark_data["Injury"].value_counts(dropna=False).head()


```

```python
shark_data.drop("pdf", axis=1)  # axis=1 refers to column-wise dropping
```

```python
# pandas returns copies, the above won't change anything, so we have to make a new variable
shark_data = shark_data.drop("pdf", axis=1)
```

```python
# Instead of dropping, why not just select the columns you want?
shark_data = shark_data[["Type", "Country", "Species ", "Injury", "Date", "Sex"]]
shark_data.columns
```

```python
# Rename columns using a dictionary
shark_data = shark_data.rename(columns={"Species ": "species", "Injury": "injury"})
shark_data.columns
```

```python
shark_data.columns = shark_data.columns.str.lower()
shark_data.columns
```

```python
# Normalise the injury column
shark_data["injury"] = shark_data["injury"].str.lower().str.strip()
shark_data["injury"].value_counts()
```

```python
mode_shark = shark_data["species"].mode().iloc[0]
shark_data["species"] = shark_data["species"].fillna(mode_shark)
```

```python
shark_data = shark_data.dropna()  # Drops all rows with na values
# Create a new column "fatal" and store True if the "injury" column is "FATAL", else False
shark_data["fatal"] = shark_data["injury"].map(lambda row: "fatal" in row)

# confirm True/False distribution looks reasonable
shark_data["fatal"].value_counts()
```

```python
shark_data["fatal"].value_counts() / len(shark_data) * 100
```
